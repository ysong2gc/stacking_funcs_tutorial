```python
#I will be using the toy model simulations to generate TS distributions to be stacked in the tutorial.
#The first line here imports all the proper functions needed from the toy model module package to generate the TS distributions. 
from toy_gaussian_star import log_like, log_like_f, get_ROI

#This is importing all simple stacking functions
from stacking_funcs import cumTS, deltaTScum, resampling_CFTS, draw_the_plot
```

```python
##This is using the toy model to create the TS distributions
##of the sources (TS) and noise (TS2)
nor = 400 #number of ROIs to look at.
nop = 317*2 #number of pixels in ROI
g = 0.03

average_counts=19.447910

f_a=2*average_counts*g/(g+2)
average_counts=2*average_counts/(g+2)
f2=average_counts/50.


TS=np.array([])
TS2=np.array([])
p_counts=np.array([])

###Constructing all ROIs and calculate log(L), then the TS value.
for i in range(0,nor):

    p_counts=get_ROI(nop, average_counts, f_a)
    pc2=get_ROI(nop, average_counts, f2)

    logL1 = log_like_f(p_counts,average_counts)
    logL2 = log_like_f(p_counts,average_counts, f_a)

    logL3 = log_like_f(pc2,average_counts)
    logL4 = log_like_f(pc2,average_counts, f2)

    TS2 = np.append(TS2, 2*(logL4-logL3))
    if i<300:
        TS = np.append(TS,2*(logL2-logL1))
```

```python
##To save time, this step is to calculate the 
##cumulative TS distribution first and save them into npy files. 
##One note is that TS values below 0 should all be discarded.
a1,a2=deltaTScum('toy_CFs',TS2, len(TS[TS>0]), 100, 1000)
b1,b2=deltaTScum('toy_sig',TS, len(TS[TS>0]), 100, 1000)
```
